> **Credit Card Fraud Detection**


```python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as pt
```

**Importing the Dataset**

```python
data = pd.read_csv("dataset\KAGGLE-creditcarddataset.csv")
data.head()
```

**Some Info about the dataset**

It contains only numerical input variables which are the result of a PCA transformation. Unfortunately, due to confidentiality issues, we cannot provide the original features and more background information about the data. Features V1, V2, … V28 are the principal components obtained with PCA, the only features which have not been transformed with PCA are 'Time' and 'Amount'. Feature 'Time' contains the seconds elapsed between each transaction and the first transaction in the dataset. The feature 'Amount' is the transaction Amount, this feature can be used for example-dependant cost-sensitive learning. Feature 'Class' is the response variable and it takes value 1 in case of fraud and 0 otherwise.

**Features and label selection**

```python
data = data.drop(columns = 'Time')
features = data.drop(columns = 'Class')
label = data['Class']
```

**Train Test  split**

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain, Xtest, Ytrain,Ytest = train_test_split(features,label,test_size = 0.2,random_state = 42)
```

```python
Xtrain.shape
```

```python
Xtest.shape
```

**Preprocessing before passing to model**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
Xtrain=sc.fit_transform(Xtrain)
Xtest=sc.transform(Xtest)
```

**Model Training and evaluation**

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model_LR = LogisticRegression()
```

```python
model_LR.fit(Xtrain, Ytrain)
LR_pred = model_LR.predict(Xtest)
```

```python
acc_LR = accuracy_score(Ytest,LR_pred)
print(acc_LR)
```

K- Nearest Neighbor Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
model_KNN = KNeighborsClassifier(n_neighbors=5, metric="euclidean")
```

```python
model_KNN.fit(Xtrain, Ytrain)
KNN_pred = model_KNN.predict(Xtest)
```

```python
acc_KNN = accuracy_score(Ytest,KNN_pred)
print(acc_KNN)
```

Support Vector Classifier

```python
from sklearn.svm import SVC
model_SC = SVC()
```

```python
model_SC.fit(Xtrain, Ytrain)
SC_pred = model_SC.predict(Xtest)
```

```python
acc_SC = accuracy_score(Ytest,SC_pred)
print(acc_SC)
```

XGBoost Classifier

```python
from xgboost import XGBClassifier
model_xbc = XGBClassifier()
```

```python
model_xbc.fit(Xtrain,Ytrain)
XB_pred = model_xbc.predict(Xtest)
```

```python
acc_XG = accuracy_score(Ytest,XB_pred)
print(acc_XG)
```

Plotting the acuuracies of all the models

```python
accuracy = [acc_LR,acc_KNN,acc_SC,acc_XG]
model = ["Linear Regression","KNN","SVC","XGBoost"]
dframe = pd.DataFrame(accuracy,model)
pt.bar(dframe,x=model,y=accuracy)
```

  **Clearly XGBoost is the Winner here with highest accuracy of 0.999631 !!**
